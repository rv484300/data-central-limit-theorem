# Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Exercise

Verify the Central Limit Theorem using the tips dataset.
Randomly select dinners and compute the mean of their tips.

```python
tips = sns.load_dataset("tips")
samples = [tips.tip.sample(30).mean() for _ in range(1000)]

plt.hist(samples, bins=30)
plt.title("Central Limit Theorem - Tips Dataset")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

print(f"Mean: {np.mean(samples):.2f}")
print(f"Std: {np.std(samples):.2f}")
```

## Comparison with Theoretical Values

According to the Central Limit Theorem:

- Sample mean ≈ Population mean
- Sample standard deviation ≈ Population std / √n

```python
population_mean = tips.tip.mean()
population_std = tips.tip.std()

print(f"Population Mean: {population_mean:.2f}")
print(f"Sample Mean (CLT): {np.mean(samples):.2f}")

print(f"Theoretical Std: {population_std/np.sqrt(30):.2f}")
print(f"Observed Std: {np.std(samples):.2f}")
```
